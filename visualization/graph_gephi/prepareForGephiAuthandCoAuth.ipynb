```python
data = researches_df[researches_df['citedby_count'] > 150]
data.head()
```

```python
from itertools import combinations

# Create an empty list to store edges
edges = []

# Iterate over each row in the dataset
for _, row in data.iterrows():
    authors = str(row['authors_id']).split('|')
    title = row['title']
    first_author = authors[0]
    # Create edges from the first author to each other author in the same publication
    for other_author in authors[1:]:
        edges.append([first_author, other_author, title])

# Convert the edge list into a DataFrame
edges_df = pd.DataFrame(edges, columns=['Source', 'Target', 'Title'])

# Save the edges to a CSV file
edges_csv_path = 'author_edges.csv'
edges_df.to_csv(edges_csv_path, index=False)

edges_df.head(), edges_csv_path
```
